```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sc
from pathlib import Path

import prepare_data
import plot
```

```python
cond = 'sel-'
cond_col = 'limegreen'
```

# Best individual over generations

```python
control_seed_dfs, control_data_dfs = prepare_data.aggregate_best_ind_stats('control')
mut_seed_dfs, mut_data_dfs = prepare_data.aggregate_best_ind_stats(cond)
```

```python
data_dfs = {}
for s in control_data_dfs:
    data_dfs[s] = pd.concat([control_data_dfs[s], mut_data_dfs[s]], axis=1)
```

## Genome size

```python
data_n = 'Genome size'
df = data_dfs[data_n]
mean_values_df = prepare_data.extract_mean_values_last_gen(df, cond)
mean_values_df
```

Evolution over generation for control and condition

```python
plot.plot_evolution_over_generation(df, data_n, ['grey']*5 + [cond_col]*5)
```

Barplot with the mean values for the last 10,000 generations

```python
plot.plot_mean_values_barplot(mean_values_df, data_n, ['grey', cond_col])
```

Barplot with the proportion of change between control and condition

```python
plot.plot_prop_barplot(mean_values_df, data_n, cond_col)
```

```python
sc.wilcoxon(mean_values_df.iloc[:,0], mean_values_df.iloc[:,1])
```
